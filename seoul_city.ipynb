```python
df_vulnerable=pd.DataFrame({'Year':['2017','2018','2019','2020','2021','2022'],
                            'Vulnerable':[1185,1239,1287,1405,1514,1572]})
```

```python
df_vulnerable=pd.DataFrame({'Year':['2017','2018','2019','2020','2021','2022'],
                            'Non-residential building':[327,319,318,316,319,314],
                            'Other than housing':[858,920,969,1089,1195,1258],
                            'total':[1185,1239,1287,1405,1514,1572]})
```

```python
palette1=sns.color_palette("GnBu_d")
```

```python
plt.rc('font',family='NanumBarunGothic')
```

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager

# 사용할 한글 폰트 경로를 지정합니다. 실제 경로로 변경하세요.
font_path = '/Library/Fonts/NanumBarunGothic.ttf'  # NanumBarunGothic.ttf 파일의 경로로 변경

# 사용할 폰트를 등록합니다.
font_manager.fontManager.addfont(font_path)

# 사용할 한글 폰트를 설정합니다.
plt.rcParams['font.family'] = 'NanumBarunGothic'
```

```python
df_vulnerable['total'] = df_vulnerable['Non-residential building'] + df_vulnerable['Other than housing']
df_long = df_vulnerable.melt(id_vars=['Year'], var_name='Category', value_name='Value')
plt.figure(figsize=(12, 6))
sns.barplot(x='Year', y='Value', hue='Category', data=df_long, palette=palette1)


plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Vulnerable Population Over the Years by Category')

plt.legend(title='Category')
plt.show()
```

# 상관관계분석

```python
df_vulnerable
```

```python
df_corr=df_vulnerable.iloc[:,[0,3]]
new_column_name = 'vulnerable'
df_corr.rename(columns={df_corr.columns[1]: new_column_name}, inplace=True)
df_corr
```

```python
data=pd.merge(df_overdose,df_GDP,on='Year',how='left')
data
```

```python
data=pd.merge(data,df_homePrice,on='Year',how='left')
data
```

```python
data=pd.merge(data,df_corr,on='Year',how='inner')
data
```

```python
correlation_matrix = data.corr()
print(correlation_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))  # 그래프 크기 설정
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")  # annot=True는 숫자 표시, cmap은 색상 맵 설정

# 그래프 제목 설정
plt.title('Correlation Heatmap')

# 그래프 표시
plt.show()
```
